```python
import sqlite3
import pandas as pd
import numpy as np

# plots & stats
import matplotlib.pyplot as plt
from scipy import stats

# machine learning
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn import metrics

# (for Part 4 regression)
import statsmodels.api as sm

# -------------------------------------------------------------------
# Connect and read all 3 tables
# -------------------------------------------------------------------
db_path = "analytics_final.db"   # change path if needed
conn = sqlite3.connect(db_path)

finstat = pd.read_sql_query("SELECT * FROM finstat;", conn)
mktdat  = pd.read_sql_query("SELECT * FROM mktdat;", conn)
textdat = pd.read_sql_query("SELECT * FROM textdat;", conn)

# Make sure fyear is integer
for df in [finstat, mktdat, textdat]:
    df['fyear'] = df['fyear'].astype(int)

# Quick sanity checks
print(finstat.head())
print(mktdat.head())
print(textdat.head())
```

```python

# Merge market data with industry from finstat
# -------------------------------------------------------------------
df1 = mktdat.merge(
    finstat[['gvkey', 'fyear', 'ffi12_desc']],
    on=['gvkey', 'fyear'],
    how='left'
)

# -------------------------------------------------------------------
# (a) Crash frequency over time
# -------------------------------------------------------------------
crash_by_year = (
    df1.groupby('fyear')['crash']
       .agg(crash_rate='mean',    # proportion of firms with a crash
            num_crash='sum',
            num_firms='count')
       .reset_index()
)

print(crash_by_year)

# Plot  you can paste this figure into the report
plt.figure()
plt.plot(crash_by_year['fyear'], crash_by_year['crash_rate'], marker='o')
plt.xlabel("Fiscal year")
plt.ylabel("Crash frequency (proportion of firms with crash)")
plt.title("Stock price crash frequency over time")
plt.tight_layout()
plt.show()

# -------------------------------------------------------------------
# (b) Crash frequency across industries (overall)
# -------------------------------------------------------------------
crash_by_industry = (
    df1.groupby('ffi12_desc')['crash']
       .agg(crash_rate='mean', num_crash='sum', num_firms='count')
       .reset_index()
       .sort_values('crash_rate', ascending=False)
)

print(crash_by_industry)

plt.figure(figsize=(8,4))
plt.bar(crash_by_industry['ffi12_desc'], crash_by_industry['crash_rate'])
plt.xlabel("FFI12 industry")
plt.ylabel("Average crash frequency")
plt.title("Average stock price crash frequency by industry")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Merge crash indicator with textual data
df2 = mktdat.merge(
    textdat[['gvkey','fyear','num_words','filesize']],
    on=['gvkey','fyear'],
    how='left'
)

# Use num_words as the length measure; drop missing
df2 = df2.dropna(subset=['num_words'])

# Define crash vs non-crash groups
length_crash   = df2.loc[df2['crash']==1, 'num_words']
length_nocrash = df2.loc[df2['crash']==0, 'num_words']

print("Mean length (crash firms): ", length_crash.mean())
print("Mean length (non-crash firms):", length_nocrash.mean())

# Two-sample t-test (Welch)
t_stat, p_value = stats.ttest_ind(
    length_crash,
    length_nocrash,
    equal_var=False,
    nan_policy='omit'
)

print(f"t-statistic: {t_stat:.3f}")
print(f"p-value:     {p_value:.4f}")

# Boxplot (no warnings)
plt.figure(figsize=(7,5))
plt.boxplot(
    [length_crash, length_nocrash],
    tick_labels=['Crash', 'No crash']
)
plt.ylabel("Number of words in annual report")
plt.title("Readability (length) and stock price crashes")
plt.tight_layout()
plt.show()

```

```python
# -------------------------------------------------------------------
# Build a single merged dataset with all required predictors
# -------------------------------------------------------------------
df3 = (
    mktdat
    .merge(finstat, on=['gvkey', 'fyear'], how='left')
    .merge(textdat, on=['gvkey', 'fyear'], how='left')
)

# Text-based variables
df3['negativity'] = df3['num_neg_words'] / df3['num_words']
df3['tone']       = (df3['num_pos_words'] - df3['num_neg_words']) / df3['num_words']

# Some extreme / invalid values may arise if num_words=0, so replace inf with NaN
df3.replace([np.inf, -np.inf], np.nan, inplace=True)

# -------------------------------------------------------------------
# Define feature sets
# -------------------------------------------------------------------
# Model 1 predictors (no text)
feat_model1 = [
    'mktcap', 'mtb', 'roe', 'lev',
    'earnvol', 'cfvol', 'salevol', 'accm',
    'beta', 'mkt_adj_return'
]

# Model 2 adds file size, tone & negativity
feat_model2 = feat_model1 + ['filesize', 'tone', 'negativity']

# Drop rows with missing crash or predictors for each model
df3_m1 = df3.dropna(subset=feat_model1 + ['crash'])
df3_m2 = df3.dropna(subset=feat_model2 + ['crash'])

# -------------------------------------------------------------------
# Train/Test split by year
# Training: fyear <= 2013
# Test:     fyear >= 2014
# -------------------------------------------------------------------
split_year = 2014

def temporal_split(df, features):
    train = df[df['fyear'] < split_year]
    test  = df[df['fyear'] >= split_year]
    X_train = train[features]
    y_train = train['crash']
    X_test  = test[features]
    y_test  = test['crash']
    return X_train, X_test, y_train, y_test

X_train1, X_test1, y_train1, y_test1 = temporal_split(df3_m1, feat_model1)
X_train2, X_test2, y_train2, y_test2 = temporal_split(df3_m2, feat_model2)

print(X_train1.shape, X_test1.shape)
print(X_train2.shape, X_test2.shape)

# -------------------------------------------------------------------
# Helper: fit logit + evaluate
# -------------------------------------------------------------------
def fit_and_evaluate_logit(X_train, y_train, X_test, y_test, model_name=""):
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('logit', LogisticRegression(max_iter=1000))
    ])
    pipe.fit(X_train, y_train)

    y_pred = pipe.predict(X_test)
    y_prob = pipe.predict_proba(X_test)[:,1]

    # Confusion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)
    acc = metrics.accuracy_score(y_test, y_pred)
    sens = metrics.recall_score(y_test, y_pred)                # TP rate
    spec = metrics.recall_score(y_test, y_pred, pos_label=0)   # TN rate
    auc  = metrics.roc_auc_score(y_test, y_prob)

    print(f"\n=== {model_name} ===")
    print("Confusion matrix:\n", cm)
    print(f"Accuracy:    {acc:.3f}")
    print(f"Sensitivity: {sens:.3f}")
    print(f"Specificity: {spec:.3f}")
    print(f"ROC AUC:     {auc:.3f}")

    return pipe

# -------------------------------------------------------------------
# Model 1: baseline (no text)
# -------------------------------------------------------------------
model1 = fit_and_evaluate_logit(
    X_train1, y_train1, X_test1, y_test1,
    model_name="Model 1  Accounting & market variables only"
)

# -------------------------------------------------------------------
# Model 2: baseline + file size + tone + negativity
# -------------------------------------------------------------------
model2 = fit_and_evaluate_logit(
    X_train2, y_train2, X_test2, y_test2,
    model_name="Model 2  Add textual variables (filesize, tone, negativity)"
)
```

```python
# Get coefficients from the logistic regression step in the pipeline
coef1 = model1.named_steps['logit'].coef_[0]
for name, c in zip(feat_model1, coef1):
    print(f"{name:15s} {c: .4f}")

coef2 = model2.named_steps['logit'].coef_[0]
for name, c in zip(feat_model2, coef2):
    print(f"{name:15s} {c: .4f}")
```

```python
# PART 4  Sentiment Change vs Future Returns (FIXED)

df4 = df3.copy()
df4 = df4.sort_values(['gvkey','fyear'])

# Compute YoY change in negativity
df4["d_negativity"] = df4.groupby("gvkey")["negativity"].diff()

# Compute next-year return
df4["lead_ret"] = df4.groupby("gvkey")["mkt_adj_return"].shift(-1)

# Clean data
df4_clean = df4.dropna(subset=["d_negativity", "lead_ret"]).copy()  # FIX HERE

print(df4_clean[["d_negativity","lead_ret"]].describe())

# ---- OLS regression ----
X = sm.add_constant(df4_clean["d_negativity"])
y = df4_clean["lead_ret"]

model_sent = sm.OLS(y, X).fit(cov_type="HC3")
print(model_sent.summary())

# ---- OPTIONAL: portfolio-sort analysis ----
df4_clean["dneg_q"] = pd.qcut(df4_clean["d_negativity"], 5, labels=False)

print(df4_clean.groupby("dneg_q")["lead_ret"].mean())
```
