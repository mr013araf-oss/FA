```python
import sqlite3
import pandas as pd
import numpy as np

# plots & stats
import matplotlib.pyplot as plt
from scipy import stats

# machine learning
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn import metrics

# (for Part 4 regression)
import statsmodels.api as sm

# -------------------------------------------------------------------
# Connect and read all 3 tables
# -------------------------------------------------------------------
db_path = "analytics_final.db"   # change path if needed
conn = sqlite3.connect(db_path)

finstat = pd.read_sql_query("SELECT * FROM finstat;", conn)
mktdat  = pd.read_sql_query("SELECT * FROM mktdat;", conn)
textdat = pd.read_sql_query("SELECT * FROM textdat;", conn)

# Make sure fyear is integer
for df in [finstat, mktdat, textdat]:
    df['fyear'] = df['fyear'].astype(int)

# Quick sanity checks
print(finstat.head())
print(mktdat.head())
print(textdat.head())
```

```python
# PART 1  Crash Descriptive Analysis

# Merge industry into market data
df1 = mktdat.merge(
    finstat[['gvkey','fyear','ffi12_desc']],
    on=['gvkey','fyear'],
    how='left'
)

# ---- (a) Crash frequency over time ----
crash_by_year = (
    df1.groupby("fyear")["crash"]
       .agg(crash_rate="mean", num_crash="sum", num_firms="count")
       .reset_index()
)

print(crash_by_year)

plt.figure(figsize=(10,4))
plt.plot(crash_by_year["fyear"], crash_by_year["crash_rate"], marker="o")
plt.title("Crash Frequency Over Time")
plt.xlabel("Fiscal Year")
plt.ylabel("Proportion of Firms with Crash")
plt.grid(True)
plt.tight_layout()
plt.show()

# ---- (b) Crash frequency across industries ----
crash_by_industry = (
    df1.groupby("ffi12_desc")["crash"]
       .agg(crash_rate="mean", num_crash="sum", num_firms="count")
       .reset_index()
       .sort_values("crash_rate", ascending=False)
)

print(crash_by_industry)

plt.figure(figsize=(10,4))
plt.bar(crash_by_industry["ffi12_desc"], crash_by_industry["crash_rate"])
plt.title("Crash Frequency by Industry")
plt.xlabel("Industry (FFI12)")
plt.ylabel("Crash Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Merge crash indicator with textual data
df2 = mktdat.merge(
    textdat[['gvkey','fyear','num_words','filesize']],
    on=['gvkey','fyear'],
    how='left'
)

# Use num_words as the length measure; drop missing
df2 = df2.dropna(subset=['num_words'])

# Define crash vs non-crash groups
length_crash   = df2.loc[df2['crash']==1, 'num_words']
length_nocrash = df2.loc[df2['crash']==0, 'num_words']

print("Mean length (crash firms): ", length_crash.mean())
print("Mean length (non-crash firms):", length_nocrash.mean())

# Two-sample t-test (Welch)
t_stat, p_value = stats.ttest_ind(
    length_crash,
    length_nocrash,
    equal_var=False,
    nan_policy='omit'
)

print(f"t-statistic: {t_stat:.3f}")
print(f"p-value:     {p_value:.4f}")

# Boxplot (no warnings)
plt.figure(figsize=(7,5))
plt.boxplot(
    [length_crash, length_nocrash],
    tick_labels=['Crash', 'No crash']
)
plt.ylabel("Number of words in annual report")
plt.title("Readability (length) and stock price crashes")
plt.tight_layout()
plt.show()
```

```python
# Features for models
feat_model1 = [
    'mktcap','mtb','roe','lev',
    'earnvol','cfvol','salevol','accm',
    'beta','mkt_adj_return'
]

feat_model2 = feat_model1 + ['filesize','tone','negativity']

# Remove missing rows
df3_m1 = df3.dropna(subset=feat_model1 + ['crash_next'])
df3_m2 = df3.dropna(subset=feat_model2 + ['crash_next'])
```

```python
split_year = 2014

def split_data(df, features):
    train = df[df["fyear"] < split_year]
    test  = df[df["fyear"] >= split_year]
    return (
        train[features], test[features],
        train["crash_next"], test["crash_next"]
    )

X_train1, X_test1, y_train1, y_test1 = split_data(df3_m1, feat_model1)
X_train2, X_test2, y_train2, y_test2 = split_data(df3_m2, feat_model2)

print(X_train1.shape, X_test1.shape)
print(X_train2.shape, X_test2.shape)
```

```python
def evaluate_logit(X_train, y_train, X_test, y_test, model_name):
    pipe = Pipeline([
        ('scaler', StandardScaler()),
        ('logit', LogisticRegression(max_iter=1000, class_weight='balanced'))
    ])
    
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    y_prob = pipe.predict_proba(X_test)[:,1]

    cm = metrics.confusion_matrix(y_test, y_pred)
    acc = metrics.accuracy_score(y_test, y_pred)
    sens = metrics.recall_score(y_test, y_pred)                  # TP rate
    spec = metrics.recall_score(y_test, y_pred, pos_label=0)     # TN rate
    auc = metrics.roc_auc_score(y_test, y_prob)

    print("\n==============================")
    print(model_name)
    print("==============================")
    print("Confusion Matrix:\n", cm)
    print("Accuracy:", acc)
    print("Sensitivity:", sens)
    print("Specificity:", spec)
    print("ROC AUC:", auc)
    
    return pipe

model1 = evaluate_logit(X_train1, y_train1, X_test1, y_test1, "MODEL 1 (No Text)")
model2 = evaluate_logit(X_train2, y_train2, X_test2, y_test2, "MODEL 2 (With Text)")
```
